## Q1. Qual vﾃｭdeo teve mais visualizaﾃｧﾃｵes?

```python
most_viewed_video = df.loc[df['view_count'].idxmax()]
print(f"O vﾃｭdeo com mais visualizaﾃｧﾃｵes ﾃｩ: {most_viewed_video['title']} com {int(most_viewed_video['view_count']):,} visualizaﾃｧﾃｵes, do canal {most_viewed_video['channel_title']}.")
```

## Q2. Qual vﾃｭdeo teve mais likes?

```python
most_liked_video = df.loc[df['like_count'].idxmax()]
print(f"O vﾃｭdeo com mais likes ﾃｩ: {most_liked_video['title']} com {int(most_liked_video['like_count']):,} likes, do canal {most_liked_video['channel_title']}.")
```

## Q3. Qual vﾃｭdeo teve mais comentﾃ｡rios?

```python
most_commented_video = df.loc[df['comment_count'].idxmax()]
print(f"O vﾃｭdeo com mais comentﾃ｡rios ﾃｩ: {most_commented_video['title']} com {int(most_commented_video['comment_count']):,} comentﾃ｡rios, do canal {most_commented_video['channel_title']}.")
```

## Q4. Em qual ano teve mais vﾃｭdeos lanﾃｧados?

```python
df.drop(columns=['publish_date'], inplace=True)
```

```python
# converter a coluna 'published_date' para o tipo datetime e extrair apenas a data

df['published_date'] = pd.to_datetime(df['published_date']).dt.date
df.head(5)
```

```python
# criar uma coluna sﾃｳ com o ano

df['year'] = pd.to_datetime(df['published_date']).dt.year
df.head(1)
```

```python
anoMaisVisto = df['year'].value_counts().sort_index()
print(f"O ano com mais vﾃｭdeos lanﾃｧados ﾃｩ: {anoMaisVisto.idxmax()} com {anoMaisVisto.max()} vﾃｭdeos.")
```
