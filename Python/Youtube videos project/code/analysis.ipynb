# Análise de Dados de Vídeos do YouTube 

Este notebook tem como objetivo analisar dados de desempenho de vídeos do YouTube para identificar padrões e insights.

**Q1. Qual vídeo teve mais visualizações?**
**Q2. Qual vídeo teve mais likes?**

## Q1. Qual vídeo teve mais visualizações?

```python
most_viewed_video = df.loc[df['view_count'].idxmax()]
print(f"O vídeo com mais visualizações é: {most_viewed_video['title']} com {int(most_viewed_video['view_count']):,} visualizações, do canal {most_viewed_video['channel_title']}.")
```

## Q2. Qual vídeo teve mais likes?

```python
most_liked_video = df.loc[df['like_count'].idxmax()]
print(f"O vídeo com mais likes é: {most_liked_video['title']} com {int(most_liked_video['like_count']):,} likes, do canal {most_liked_video['channel_title']}.")
```

## Q3. Qual vídeo teve mais comentários?

```python
most_commented_video = df.loc[df['comment_count'].idxmax()]
print(f"O vídeo com mais comentários é: {most_commented_video['title']} com {int(most_commented_video['comment_count']):,} comentários, do canal {most_commented_video['channel_title']}.")
```

## Q4. Em qual ano teve mais vídeos lançados?

```python
df.drop(columns=['publish_date'], inplace=True)
```

```python
# converter a coluna 'published_date' para o tipo datetime e extrair apenas a data

df['published_date'] = pd.to_datetime(df['published_date']).dt.date
df.head(5)
```

```python
# criar uma coluna só com o ano

df['year'] = pd.to_datetime(df['published_date']).dt.year
df.head(1)
```

```python
anoMaisVisto = df['year'].value_counts().sort_index()
print(f"O ano com mais vídeos lançados é: {anoMaisVisto.idxmax()} com {anoMaisVisto.max()} vídeos.")
```
