```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_excel(r'../datasets/Harassment-Bullying-on-basis-of-sex_reported.xlsx', sheet_name='Total', header=4)
df.head(5)
```

```python
# Remover linhas totalmente vazias
df_clean = df.dropna(how='all')

# Remover colunas totalmente vazias
df_clean = df_clean.dropna(axis=1, how='all')

# Remover as primeiras linhas de cabeçalho extras (exemplo: linhas 0 e 1)
df_clean = df_clean.iloc[2:]

# Redefinir o índice
df_clean = df_clean.reset_index(drop=True)

# Renomear as colunas para facilitar o uso
df_clean.columns = [
    'Description', 'State', 'Total', 
    'AIAN_Number', 'AIAN_Percent', 'Asian_Number', 'Asian_Percent',
    'Hispanic_Number', 'Hispanic_Percent', 'Black_Number', 'Black_Percent',
    'White_Number', 'White_Percent', 'NHPI_Number', 'NHPI_Percent',
    'TwoOrMore_Number', 'TwoOrMore_Percent', 'SDISABI_Number', 'SDISABI_Percent', 'SDISABI504_Number', 'EngLearn_Number', 'SDISABI504_Percent',
    'EngLearn_Percent',   'Schools_Number', 'Schools_Percent_Reported'
][:len(df_clean.columns)]  # Garante que não exceda o número de colunas

# Exibir as 5 primeiras linhas limpas
df_clean.head()
```

```python
df_clean = df_clean.drop('Description', axis=1)  # Remover a coluna 'Description'
df_clean.head(5)
```

```python
df_clean['Schools_Percent_Reported'] = df_clean['Schools_Percent_Reported'].round(2)
df_clean.head(5)
```

```python
df_clean.info()
```

```python
def floatTransform(value):
    try:
        return float(value)
    except (ValueError, TypeError):
        return value

for col in df_clean.columns:
    df_clean[col] = df_clean[col].apply(floatTransform)

df_clean.info()
```

```python
df_clean.head(5)
```

```python
def roundValues(value):
    try:
        return round(value, 2)
    except (ValueError, TypeError):
        return value

numericCols = df_clean.select_dtypes(include=['number']).columns

for col in numericCols:
    df_clean[col] = df_clean[col].apply(roundValues)

df_clean.head(5)
```
