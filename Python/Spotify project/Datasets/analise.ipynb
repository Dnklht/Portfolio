```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as mlt
import chardet # Detectar a codificação do dataset
import openpyxl
```

```python
with open("Most Streamed Spotify Songs 2024.csv", 'rb') as file:
    raw_data = file.read()
    result = chardet.detect(raw_data)
    encoding = result['encoding']
    print(f"O enconding do dataset é: {encoding}")
```

```python
spotify = pd.read_csv("Most Streamed Spotify Songs 2024.csv", encoding=result['encoding'])
pd.set_option('display.max_columns', None)
spotify.head(20)
```

```python
for i, col in enumerate(spotify.columns):
    print(f"{i + 1}: {col}")
```

```python
isna = spotify.isna().sum()
isna
```

```python
sorted_isna = isna.sort_values(ascending=False)
sorted_isna
```

```python
spotify = spotify.fillna('NA')
spotify
```

```python
for col in spotify.columns:
    na_values = spotify[col].isna().sum()
    print(f"A coluna '{col}' tem {na_values} nulos")
```

```python
def separation_year(coluna):
    return int(coluna.split('/')[-1])

spotify['Release Year'] = spotify['Release Date'].apply(separation_year)
spotify['Release Year']
```

```python
spotify.head(20)
(spotify['Spotify Streams'] == 'NA').sum()
```

```python
spotify = spotify[spotify['Spotify Streams'].apply(lambda x: x != 'NA')]
spotify
```

```python
is_NA = (spotify['Spotify Streams'] == 'NA').sum()
is_NA
```

```python
spotify = spotify.sort_values(by='Spotify Streams', ascending=False)
spotify
```

```python
spotify['Release Date'] = pd.to_datetime(spotify['Release Date'], dayfirst=True, errors='coerce')

# Formatando a coluna para o formato 'DD/MM/YYYY'
spotify['Release Date'] = spotify['Release Date'].dt.strftime('%d/%m/%Y')
spotify
```

```python
spotify.columns
```

```python
cols = ['ISRC', 'Pandora Streams', 'Pandora Track Stations', 'Explicit Track', 'AirPlay Spins', 'SiriusXM Spins', 'TIDAL Popularity']

spotify = spotify.drop(columns=cols)
spotify
```

```python
gb_year = spotify.groupby(by='Release Year')['Track'].count()
gb_year_sorted = gb_year.sort_values(ascending=False).head(10)
```

```python
for year, count in gb_year_sorted.items():
    print(f'No ano de {year}, foi lançado {count} músicas.')
```

```python
spotify = spotify.dropna(subset=['Release Date'])
spotify
```
